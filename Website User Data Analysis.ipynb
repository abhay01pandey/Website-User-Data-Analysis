# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load and Preview Data

```python
df = pd.read_csv("data-export (1).csv")
```

```python
df.head()
```

# Rename Columns

```python
df.columns = df.iloc[0]
df = df.drop(index = 0).reset_index(drop=True)
df.columns = ["channel group", "Datehour","users", "Sessions", "Engaged Sessions", "Average engagement time per session", "Engaged sessions per user","Events per session","Engagement rate","Event count"]
```

```python
df.head()
```

#  Data Info and Overview

```python
df.info()
```

# Convert Data Types

```python
df["Datehour"]= pd.to_datetime(df["Datehour"],format="%Y%m%d%H",errors='coerce')
```

```python
numeric_cols= df.columns.drop(["channel group","Datehour"])
df[numeric_cols]= df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df["Hour"]= df["Datehour"].dt.hour
```

```python
df.describe()
```

# Set Plot Style

```python
sns.set(style ="whitegrid")
```

# Visualization: Sessions & Users Over Time

```python
plt.figure(figsize=(10,5))
df.groupby("Datehour")[["Sessions","users"]].sum().plot(ax=plt.gca())
plt.title("Sessions & users over time")
plt.xlabel("Datehour")
plt.ylabel("count")
plt.show()
```

# Visualization: Total Users by Channel

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="channel group", y="users",
            estimator=np.sum, hue="channel group", legend=False, palette="viridis")
plt.title("Total users by channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

# Visualization: Average Engagement Time by Channel

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x="channel group",
    y="Average engagement time per session",
    estimator=np.mean,
    hue="channel group",
    legend=False,
    palette="magma"
)
plt.title("Avg Engagement Time by channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

# Visualization: Engagement Rate Distribution by Channel

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="channel group", y="Engagement rate",
            hue="channel group", palette="coolwarm", legend=False)
plt.title("Engagement Rate Distribution by channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

# Visualization: Traffic by Hour & Channel

```python
heatmap_data=df.groupby(["Hour","channel group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data,cmap="YlGnBu",linewidth=5,annot=True,fmt='.0f')
plt.title("Traffic by Hour & Channel")
plt.xlabel("channel Group")
plt.ylabel("Hour of Day")
plt.show()
```
